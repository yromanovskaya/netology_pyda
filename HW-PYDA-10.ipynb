```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1 = pd.read_csv('tmdb_5000_movies.csv')
df_1.head()
```

```python
df_1.info()
```

```python
# Удалим из датафрейма строки с путсыми значениями интересующих нас признаков, а именно:
df = df_1.dropna(subset=['runtime', 'release_date'])
df.info()
# Индексы не переписываем, оставляем соответствующими первоначальному датафрейму
```

# 1. Линейный график

Посмотрим, как много фильмов из анализируемого датафрейма выходило ежегодно, для этого выделим из колонки release_date год выпуска. 
Найдем через группировку, сколько фильмов выходило ежегодно.
Посторим линейный график зависимости года (x) от количества вышедших фильмов (y)

```python
def year(a):
    return int(a[:4])
df['year']=df['release_date'].apply(year)
# df.head()
df_year = df.groupby('year')['id'].count()
# df_year.tail(20)
df_year.plot()
```

По графику видно, что примерно с 1960-х появилась тенденция к увеличению выпускаемых фильмов, в 90-е был скачок динамики выхода фильмов по количеству

Посмотрим, как изменялся бюджет и доход от релизов фильмов во времени. 
Для этого используем данные колонок budget, revenue, year. При этом возьмем средние годовые значения доходов и бюджетов.
Построим два линейных графика для более удобного сравнения - year(x), budget, revenue (y)

```python
df_year_1 = df.groupby('year')['budget','revenue'].mean()
# df_year_1.head()
df_year_1.plot()
```

Анализируя график,можем предположить, что: 
1. в данных датасета присутствует вылет в признаке budget (в конце 1920-х вряд ли выходил фильм с бюджетом 90 млрд.у.е.)
2. Бюджеты фильмов с течением времени медленно ростут на протяжении всего анализируемого периода. При этом наблюдаем, что в период 1980-2000 рост бюджетов ускорился по сравнению с анализируемым периодом
3. Доход от выпуска фильмов также на всем протяжении анализируемого периода в среднем растет. Скачки графика вызваны скорее не периодической зависимсотью, а выходами культовых фильмов. 
4. Доход от фильмов стал активнее расти раньше скачка роста бюджетов, где-то с 1970-х. Скорее всего этим и вызван скачок роста бюджетов чуть позже. 
5. Рост доходов и бюджетов фильмов на протяжении анализируемого периода имеет положительную динамику в обоих случаях, при этом рост доходов с проката круче, счем рост бюджетов.

# 3. Круовая диаграмма

Понимаем, что скорее всегод в датасете большее количество фильмов на английском языке. Проверим это. Сгруппируем датасет по колонке 'original_language', применим к остальным суммирование количества значений, тут же отсортируем его от большего значения к меньшему и выведем график только по самым часто встречающимся языкам.

```python
df_pie = df.groupby(['original_language']).count().sort_values(by='id',ascending=False).head(5)
# df_pie
df_pie.plot(kind='pie', y='id')
```

С помощью групповой диаграммы визуально подтвердили, что подавляющее большинство фильмв выходит на английском языке

# 5. Корреляция

Проанавлизируем взаимосвязь  количества оценок и популярности фильмов. 

```python
plt.scatter(df['popularity'], df['vote_count'])
```

```python
sns.jointplot(x='popularity', y='vote_count', 
              data=df, kind='reg')
```

Уберем фильмы с популярностью выше 250 для более комфортной визуализации скопления данных в пределах интервала популярности от0 до 250.

```python
sns.jointplot(x='popularity', y='vote_count', 
              data=df.query('popularity <= 250'), kind='reg')
```

Глядя на график, можем сказать, что количетсво оценок и популярность взаимосвязаны - чем больше оценок, тем выше показатель популярности.
